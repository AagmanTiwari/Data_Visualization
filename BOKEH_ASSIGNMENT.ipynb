```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Set up Bokeh to output the plot inline in the notebook
output_notebook()

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the results
show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
# markers based on the 'sizes' and 'colors' columns.
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import output_file
from bokeh.models import ColumnDataSource
import numpy as np

# Set up Bokeh to output the plot inline in the notebook
output_notebook()

# Generate random data
np.random.seed(42)
x = np.random.rand(100) * 100
y = np.random.rand(100) * 100
sizes = np.random.rand(100) * 30 + 10  # Sizes ranging from 10 to 40
colors = np.random.randint(0, 255, 100)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot
p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='x', y_axis_label='y')

# Add scatter renderer
p.scatter('x', 'y', size='sizes', color='colors', fill_alpha=0.6, line_color=None, source=source, legend_label="Random Points")

# Show the results
show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral6

# Set up Bokeh to output the plot inline in the notebook
output_notebook()

# Given dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot
p = figure(x_range=fruits, title="Counts of Different Fruits", x_axis_label='Fruits', y_axis_label='Counts')

# Add a bar renderer
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the results
show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)
```

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Set up Bokeh to output the plot inline in the notebook
output_notebook()

# Generate the data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Histogram of Data Distribution", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad renderer
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the results
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
# data_heatmap = np.random.rand(10, 10)
# x = np.linspace(0, 1, 10)
# y = np.linspace(0, 1, 10)
# xx, yy = np.meshgrid(x, y)
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import transform
from bokeh.io import output_notebook
import numpy as np

# Set up Bokeh to output the plot inline in the notebook
output_notebook()

# Generate the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a new plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', x_range=(0, 1), y_range=(0, 1))

# Use the 'image' glyph to display the heatmap
color_mapper = LinearColorMapper(palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the results
show(p)
```
