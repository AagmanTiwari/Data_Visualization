```python
# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic 
# dataset.
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(0)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5

# Create a DataFrame for easier plotting with Seaborn
import pandas as pd
data = pd.DataFrame({'x': x, 'y': y})

# Create scatter plot
sns.scatterplot(x='x', y='y', data=data)

# Add titles and labels
plt.title('Scatter Plot of Synthetic Dataset')
plt.xlabel('X Variable')
plt.ylabel('Y Variable')

# Show the plot
plt.show()
```

```python
# 2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate random dataset
np.random.seed(0)
data = np.random.randn(1000)  # 1000 data points with a normal distribution

# Create a DataFrame for easier plotting with Seaborn
import pandas as pd
df = pd.DataFrame({'data': data})

# Visualize the distribution with a histogram and KDE plot
sns.histplot(df['data'], kde=True, color='blue', bins=30)

# Add titles and labels
plt.title('Distribution of Numerical Variable')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# 3. Create a dataset representing categories and their corresponding values. Compare different categories 
# based on numerical values.
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [25, 40, 30, 35, 20]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Create a bar plot
sns.barplot(x='Category', y='Values', data=df, palette='viridis')

# Add titles and labels
plt.title('Comparison of Categories Based on Values')
plt.xlabel('Category')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical 
# variable across different categories
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.rand(100)

# Create categories for each value
cat = np.random.choice(categories, 100)

# Convert to DataFrame
df = pd.DataFrame({'Category': cat, 'Values': values})

# Visualize the distribution with a box plot
sns.boxplot(x='Category', y='Values', data=df, palette='viridis')

# Add titles and labels
plt.title('Distribution of Numerical Variable Across Categories')
plt.xlabel('Category')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a 
# heatmap.
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate synthetic dataset with correlated features
np.random.seed(0)
data_size = 100
feature_1 = np.random.randn(data_size)
feature_2 = feature_1 + np.random.randn(data_size) * 0.5
feature_3 = 2 * feature_1 - feature_2 + np.random.randn(data_size) * 0.3

# Create a DataFrame
df = pd.DataFrame({'Feature 1': feature_1, 'Feature 2': feature_2, 'Feature 3': feature_3})

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)

# Add titles
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```
